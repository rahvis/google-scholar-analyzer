```python
import json
import numpy as np
from dataclasses import dataclass
from typing import Dict, List

@dataclass
class AuthorMetrics:
    citation_spike_index: float
    self_citation_rate: float
    citation_vitality: float
    publication_gap_index: float
    quality_drop_metric: float
    activity_decline_index: float
    collaborator_dependency: float
    independence_metric: float
    leadership_index: float
    composite_score: float

class ProfileAnalyzer:
    def __init__(self, data: Dict):
        self.data = data
        self.field_adjustments = {
            "MATHEMATICS": 0.8,
            "COMPUTER_SCIENCE": 0.8,
            "PHYSICS": 1.0,
            "BIOLOGY": 1.2,
            "MEDICINE": 1.5,
            "SOCIAL_SCIENCES": 0.7
        }

    def calculate_citation_spike_index(self, author: Dict) -> float:
        citations = list(author["citations"].values())
        if len(citations) < 2:
            return 0

        current = citations[-1]
        previous = np.mean(citations[:-1])
        std = np.std(citations[:-1]) if len(citations) > 2 else 1
        return (current - previous) / (std if std > 0 else 1)

    def calculate_self_citation_rate(self, author: Dict) -> float:
        total_citations = sum(author["citations"].values())
        self_citations = sum(author["self_citations"].values())
        return (self_citations / total_citations) * 100 if total_citations > 0 else 0

    def calculate_citation_vitality(self, author: Dict) -> float:
        recent_citations = sum(list(author["citations"].values())[-2:])
        total_citations = sum(author["citations"].values())
        return recent_citations / total_citations if total_citations > 0 else 0

    def calculate_publication_gap_index(self, author: Dict) -> float:
        pubs = list(author["publications"].values())
        if len(pubs) < 2:
            return 0
        current_gap = 1 / (pubs[-1] + 0.1)  # Adding 0.1 to avoid division by zero
        avg_gap = 1 / (np.mean(pubs[:-1]) + 0.1)
        return current_gap / avg_gap

    def calculate_quality_drop_metric(self, author: Dict) -> float:
        citations_per_pub = [
            c / (p + 0.1) for c, p in zip(
                author["citations"].values(),
                author["publications"].values()
            )
        ]
        recent = np.mean(citations_per_pub[-2:])
        peak = max(citations_per_pub)
        return recent / peak if peak > 0 else 0

    def calculate_activity_decline_index(self, author: Dict) -> float:
        recent_pubs = sum(list(author["publications"].values())[-2:])
        avg_pubs = np.mean(list(author["publications"].values()))
        return recent_pubs / (2 * avg_pubs) if avg_pubs > 0 else 0

    def calculate_collaborator_dependency(self, author: Dict) -> float:
        return author["collaborators"]["most_frequent"] / sum(author["publications"].values())

    def calculate_independence_metric(self, author: Dict) -> float:
        career_weights = {
            "JUNIOR": 0.5,
            "MID_CAREER": 1.0,
            "SENIOR": 1.5
        }
        years_active = author["years_active"]
        career_stage = "JUNIOR" if years_active < 7 else "MID_CAREER" if years_active < 15 else "SENIOR"
        alpha = career_weights[career_stage]

        total_papers = sum(author["publications"].values())
        if total_papers == 0:
            return 0

        return (author["authorship_positions"]["first"] +
                alpha * author["authorship_positions"]["last"]) / total_papers

    def calculate_leadership_index(self, author: Dict) -> float:
        recent_papers = sum(list(author["publications"].values())[-5:])
        if recent_papers == 0:
            return 0
        return author["authorship_positions"]["last"] / recent_papers

    def calculate_composite_score(self, metrics: AuthorMetrics) -> float:
        weights = {
            "citation_spike_index": 0.15,
            "self_citation_rate": 0.15,
            "citation_vitality": 0.1,
            "publication_gap_index": 0.1,
            "quality_drop_metric": 0.1,
            "activity_decline_index": 0.1,
            "collaborator_dependency": 0.1,
            "independence_metric": 0.1,
            "leadership_index": 0.1
        }

        score = 0
        for metric_name, weight in weights.items():
            metric_value = getattr(metrics, metric_name)
            # Convert metrics so that higher values are better
            if metric_name in ["self_citation_rate", "publication_gap_index"]:
                metric_value = 1 - metric_value
            score += weight * metric_value

        return score

    def analyze_author(self, author: Dict) -> AuthorMetrics:
        field_adjustment = self.field_adjustments.get(author["field"], 1.0)

        metrics = AuthorMetrics(
            citation_spike_index=self.calculate_citation_spike_index(author) * field_adjustment,
            self_citation_rate=self.calculate_self_citation_rate(author) * field_adjustment,
            citation_vitality=self.calculate_citation_vitality(author) * field_adjustment,
            publication_gap_index=self.calculate_publication_gap_index(author) * field_adjustment,
            quality_drop_metric=self.calculate_quality_drop_metric(author) * field_adjustment,
            activity_decline_index=self.calculate_activity_decline_index(author) * field_adjustment,
            collaborator_dependency=self.calculate_collaborator_dependency(author) * field_adjustment,
            independence_metric=self.calculate_independence_metric(author) * field_adjustment,
            leadership_index=self.calculate_leadership_index(author) * field_adjustment,
            composite_score=0
        )

        metrics.composite_score = self.calculate_composite_score(metrics)
        return metrics

def main():
    # Load data from authors.json
    with open('/content/authors.json', 'r') as f:
        data = json.load(f)

    analyzer = ProfileAnalyzer(data)
    results = []

    # Analyze each author
    for author in data["authors"]:
        metrics = analyzer.analyze_author(author)
        results.append({
            "name": author["name"],
            "field": author["field"],
            "years_active": author["years_active"],
            "metrics": metrics,
            "composite_score": metrics.composite_score
        })

    # Sort by composite score
    results.sort(key=lambda x: x["composite_score"], reverse=True)

    # Print rankings
    print("\nScholar Rankings:")
    print("=" * 50)
    for i, result in enumerate(results, 1):
        print(f"\n{i}. {result['name']} ({result['field']})")
        print(f"   Years Active: {result['years_active']}")
        print(f"   Composite Score: {result['metrics'].composite_score:.3f}")
        print(f"   Key Metrics:")
        print(f"   - Citation Vitality: {result['metrics'].citation_vitality:.2f}")
        print(f"   - Quality Drop Metric: {result['metrics'].quality_drop_metric:.2f}")
        print(f"   - Independence Metric: {result['metrics'].independence_metric:.2f}")

        # Identify red flags
        red_flags = []
        if result['metrics'].self_citation_rate > 30:
            red_flags.append("High self-citation rate")
        if result['metrics'].publication_gap_index > 2:
            red_flags.append("Significant publication gaps")
        if result['metrics'].collaborator_dependency > 0.4:
            red_flags.append("High collaborator dependency")

        if red_flags:
            print(f"   Red Flags: {', '.join(red_flags)}")

if __name__ == "__main__":
    main()
```

```python
import json
import networkx as nx
import plotly.graph_objects as go
from dash import Dash, dcc, html, Input, Output

# Load authors data
with open('/content/authors.json', 'r') as f:
    authors_data = json.load(f)["authors"]

# Generate rankings based on composite scores
def calculate_composite_score(author):
    # Example formula for composite score, modify as needed
    citations = sum(author["citations"].values())
    publications = sum(author["publications"].values())
    self_citation_rate = sum(author["self_citations"].values()) / (citations + 1)
    collaborator_dependency = author["collaborators"]["most_frequent"] / (publications + 1)
    return (citations * 0.4) - (self_citation_rate * 20) - (collaborator_dependency * 10)

rankings = sorted(authors_data, key=calculate_composite_score, reverse=True)
for rank, author in enumerate(rankings, start=1):
    author["rank"] = rank
    author["composite_score"] = calculate_composite_score(author)

# Create a graph visualization
G = nx.DiGraph()
for author in rankings:
    G.add_node(author["name"], rank=author["rank"], composite_score=author["composite_score"])

# Create Dash app
app = Dash(__name__)

app.layout = html.Div([
    html.H1("Author Rankings", style={'textAlign': 'center'}),
    dcc.Graph(id="graph"),
    html.Div(id="author-details", style={"marginTop": "20px", "padding": "10px"})
])

@app.callback(
    [Output("graph", "figure"), Output("author-details", "children")],
    [Input("graph", "clickData")]
)
def update_graph(click_data):
    clicked_author = None
    if click_data:
        clicked_author = click_data["points"][0]["text"]

    # Create graph figure
    pos = nx.spring_layout(G)
    edge_x, edge_y, node_x, node_y, text, size = [], [], [], [], [], []
    for edge in G.edges():
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_x.append(x0)
        edge_y.append(y0)
        edge_x.append(x1)
        edge_y.append(y1)
        edge_x.append(None)
        edge_y.append(None)
    edge_trace = go.Scatter(x=edge_x, y=edge_y, line=dict(width=0.5, color="#888"),
                            hoverinfo="none", mode="lines")

    for node in G.nodes():
        x, y = pos[node]
        node_x.append(x)
        node_y.append(y)
        size.append(20 if node == clicked_author else 10)
        text.append(node)
    node_trace = go.Scatter(
        x=node_x, y=node_y, mode="markers+text", hoverinfo="text",
        marker=dict(size=size, color="LightSkyBlue", line=dict(width=2)),
        text=text
    )

    fig = go.Figure(data=[edge_trace, node_trace],
                    layout=go.Layout(
                        title="Author Graph",
                        titlefont_size=16,
                        showlegend=False,
                        hovermode="closest",
                        margin=dict(b=0, l=0, r=0, t=40),
                        xaxis=dict(showgrid=False, zeroline=False),
                        yaxis=dict(showgrid=False, zeroline=False)
                    ))

    # Display details for clicked author
    author_details = ""
    if clicked_author:
        author = next(a for a in rankings if a["name"] == clicked_author)
        metrics = {
            "Rank": author["rank"],
            "Composite Score": round(author["composite_score"], 2),
            "Citations": sum(author["citations"].values()),
            "Publications": sum(author["publications"].values()),
            "Self-citation Rate": round(sum(author["self_citations"].values()) / sum(author["citations"].values()) * 100, 2)
        }
        author_details = html.Div([
            html.H3(f"Details for {clicked_author}"),
            html.Ul([html.Li(f"{key}: {value}") for key, value in metrics.items()])
        ])

    return fig, author_details

if __name__ == "__main__":
    app.run_server(debug=True)
```

```python
import json
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

# Load data
with open("/content/authors.json", "r") as f:
    data = json.load(f)

# Function to calculate composite score
def calculate_composite_score(metrics):
    weights = {
        "citation_spike_index": 0.15,
        "self_citation_rate": 0.15,
        "citation_vitality": 0.1,
        "publication_gap_index": 0.1,
        "quality_drop_metric": 0.1,
        "activity_decline_index": 0.1,
        "collaborator_dependency": 0.1,
        "independence_metric": 0.1,
        "leadership_index": 0.1
    }
    score = 0
    for key, weight in weights.items():
        value = metrics[key]
        # Convert values for metrics where lower values are better
        if key in ["self_citation_rate", "publication_gap_index"]:
            value = 1 - value
        score += weight * value
    return score

# Function to calculate author metrics
def calculate_author_metrics(author):
    metrics = {
        "citation_spike_index": 0.5,  # Placeholder value for this example
        "self_citation_rate": 0.25,  # Placeholder value for this example
        "citation_vitality": 0.8,  # Placeholder value for this example
        "publication_gap_index": 1.2,  # Placeholder value for this example
        "quality_drop_metric": 0.3,  # Placeholder value for this example
        "activity_decline_index": 0.6,  # Placeholder value for this example
        "collaborator_dependency": 0.2,  # Placeholder value for this example
        "independence_metric": 0.7,  # Placeholder value for this example
        "leadership_index": 0.4  # Placeholder value for this example
    }
    return metrics

# Function to analyze and rank authors based on composite score
def analyze_authors(data):
    authors = []
    for author in data["authors"]:
        metrics = calculate_author_metrics(author)  # Calculate metrics for each author
        composite_score = calculate_composite_score(metrics)  # Calculate composite score

        authors.append({
            "name": author["name"],
            "rank": 0,  # Placeholder, will assign later
            "composite_score": composite_score,
            "metrics": metrics
        })

    # Rank authors by composite score (descending)
    authors.sort(key=lambda x: x["composite_score"], reverse=True)
    for i, author in enumerate(authors, 1):
        author["rank"] = i
    return authors

# Graph creation with hover interaction
def create_graph(authors):
    # Create a graph
    G = nx.Graph()
    labels = {}

    # Add nodes with rank-prefixed labels
    for author in authors:
        node_label = f"Rank {author['rank']} {author['name']}"
        G.add_node(author["name"], label=node_label, data=author)
        labels[author["name"]] = node_label

    # Layout the graph
    pos = nx.spring_layout(G)

    # Draw the graph
    fig, ax = plt.subplots(figsize=(10, 8))
    nx.draw_networkx_nodes(G, pos, node_size=2000, node_color="skyblue", ax=ax)
    nx.draw_networkx_labels(G, pos, labels, font_size=10, font_color="black", ax=ax)
    nx.draw_networkx_edges(G, pos, alpha=0.3, ax=ax)

    # Hover functionality
    def on_hover(event):
        if event.inaxes == ax:
            for node, (x, y) in pos.items():
                dx = event.xdata - x
                dy = event.ydata - y
                dist = (dx ** 2 + dy ** 2) ** 0.5
                if dist < 0.05:  # Adjust sensitivity
                    author_data = G.nodes[node]["data"]
                    tooltip_text = (
                        f"{labels[node]}\n"
                        f"Composite Score: {author_data['composite_score']:.2f}\n"
                        f"Metrics:\n"
                        f"  - Citation Spike Index: {author_data['metrics']['citation_spike_index']:.2f}\n"
                        f"  - Self-Citation Rate: {author_data['metrics']['self_citation_rate']:.2f}%\n"
                        f"  - Citation Vitality: {author_data['metrics']['citation_vitality']:.2f}\n"
                        f"  - Publication Gap Index: {author_data['metrics']['publication_gap_index']:.2f}\n"
                        f"  - Quality Drop Metric: {author_data['metrics']['quality_drop_metric']:.2f}"
                    )

                    # Show tooltip
                    ax.add_patch(
                        FancyBboxPatch(
                            (event.xdata + 0.05, event.ydata + 0.05),
                            0.4,
                            0.3,
                            boxstyle="round,pad=0.3",
                            edgecolor="black",
                            facecolor="white",
                            alpha=0.8
                        )
                    )
                    ax.text(
                        event.xdata + 0.07,
                        event.ydata + 0.25,
                        tooltip_text,
                        fontsize=9,
                        wrap=True,
                    )
                    plt.draw()
                    return

    fig.canvas.mpl_connect("motion_notify_event", on_hover)
    plt.axis("off")
    plt.show()

# Main execution
def main():
    authors = analyze_authors(data)
    create_graph(authors)

if __name__ == "__main__":
    main()
```

```python
import random
import json
from faker import Faker

# Initialize Faker to generate random names
fake = Faker()

# Function to generate random values for a given year range
def generate_random_data(start_year, end_year, num_years):
    citations = {}
    self_citations = {}
    publications = {}

    for year in range(start_year, end_year + 1):
        citations[year] = random.randint(30, 300)  # Random citations between 30 and 300
        self_citations[year] = random.randint(5, 100)  # Random self-citations between 5 and 100
        publications[year] = random.randint(3, 15)  # Random publications between 3 and 15

    return citations, self_citations, publications

# Function to generate random author data
def generate_author(start_year, end_year):
    years_active = end_year - start_year + 1
    citations, self_citations, publications = generate_random_data(start_year, end_year, years_active)

    # Generate random name using Faker
    name = fake.name()

    author = {
        "name": name,
        "field": "COMPUTER_SCIENCE",
        "years_active": years_active,
        "citations": citations,
        "self_citations": self_citations,
        "publications": publications,
        "collaborators": {
            "most_frequent": random.randint(2, 6)  # Random number of most frequent collaborators
        },
        "authorship_positions": {
            "first": random.randint(10, 30),
            "last": random.randint(5, 20)
        }
    }

    return author

# Function to generate the synthetic dataset for multiple authors
def generate_synthetic_data(num_authors=500, start_year=2015, end_year=2022):
    authors = []
    for _ in range(num_authors):
        author = generate_author(start_year, end_year)
        authors.append(author)

    return {"authors": authors}

# Generate the synthetic dataset
data = generate_synthetic_data(num_authors=500, start_year=2015, end_year=2022)

# Save the data to a .json file
with open("synthetic_authors_500.json", "w") as f:
    json.dump(data, f, indent=4)

print("Synthetic dataset with 500 authors has been saved to 'synthetic_authors_500.json'.")
```
